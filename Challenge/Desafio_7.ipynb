### Opción 1  (eliminar lineas)

```python
# Definición de la función censura1 que toma un nombre de archivo como parámetro
def censura1(archivo):
    # Importación de la biblioteca regex
    import regex as re
    
    # Obtención de la palabra a censurar desde la entrada del usuario
    palabra = input("¿Qué palabra no es digna de estar en el archivo nuevo?")
    
    # Apertura del archivo original con la extensión .txt y codificación utf-8
    with open(archivo + ".txt", encoding="utf-8") as archivo:
        # Inicialización de variables
        texto = ""
        cont = 1
        
        # Iteración a través de cada línea en el archivo original
        for linea in archivo:
            # Verificación si la palabra a censurar está en la línea
            if palabra in linea:
                # Imprimir mensaje indicando que la palabra está censurada en esa línea
                print(f"Palabra censurada en línea {cont}")
                # Agregar información a la variable 'texto'
                texto = texto + (f"Palabra censurada en línea {cont}\n\n")
            else:
                # Si la palabra no está en la línea, agregar la línea completa a 'texto'
                texto = texto + linea
                # Imprimir la línea sin censura
                print(linea)
            
            # Incremento del contador de líneas
            cont += 1
    
    # Escritura del texto resultante en un nuevo archivo llamado "new_censured_2.txt" con codificación utf-8
    with open("new_censured_2.txt", "w", encoding="utf-8") as new:
        new.write(texto)
    
    # Devolución del texto resultante
    return texto
```

```python
censura1("la_rosa")
```

### Opción 2  (censurar palabras)

```python
# Definición de la función censura2 que toma un nombre de archivo como parámetro
def censura2(archivo):
    # Importación de la biblioteca regex
    import regex as re
    
    # Obtención de la palabra a censurar desde la entrada del usuario
    palabra = input("¿Qué palabra no es digna de estar en el archivo nuevo?")
    
    # Apertura del archivo original con la extensión .txt y codificación unicode_escape
    with open(archivo + ".txt", encoding="utf-8") as archivo:
        # Inicialización de variables
        texto = ""
        
        # Iteración a través de cada línea en el archivo original
        for linea in archivo:
            # Utilización de regex para sustituir la palabra censurada por "@%%$&%%" en la línea
            texto = texto + re.sub(palabra, "@%%$&%%", linea)
    
    # Escritura del texto resultante en un nuevo archivo llamado "new_censured_2.txt" con codificación utf-8
    with open("new_censured_1.txt", "w", encoding="utf-8") as new:
        new.write(texto)
    
    # Devolución del texto resultante
    return texto
```
