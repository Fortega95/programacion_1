Crea un script que lea un archivo de texto y cuente el número de ocurrencias de cada palabra en el archivo. Luego, deberá escribir estas estadísticas en un nuevo archivo.

```python
# Definición de la función contador que toma un nombre de archivo como parámetro
def contador(nombre):
    
    # Importación de las bibliotecas necesarias
    import regex as rx
    import pandas as pd
    
    # Mensaje informativo
    print("Conteos de palabras en el texto:")
    
    # Lectura del archivo especificado y almacenamiento en la variable 'archivo'
    archivo = open(nombre, encoding="unicode_escape").read()
    
    # Utilización de regex para encontrar todas las palabras en el texto y convertirlas a minúsculas
    palabras = list(filter(None, [x.lower() for x in rx.findall("\w*", archivo)]))
    
    # Creación de un diccionario para almacenar la frecuencia de cada palabra
    frec = {}
    for i in list(set(palabras)):
        frec[i] = palabras.count(i)
        
    # Ordenamiento del diccionario por claves y almacenamiento en 'freq'
    freq = dict(sorted(frec.items()))
    
    # Escritura de las estadísticas en un archivo de texto llamado "estadistica.txt"
    with open("estadisticas_ejercicio6.txt", "w") as estad:
        for palabra, valor in freq.items():
            estad.write("%s %s\n" % (palabra, valor))
    
    # Devolución del diccionario de frecuencias
    return freq
```

```python
contador("la_rosa.txt")
```
